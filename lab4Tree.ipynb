```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.43, random_state=42)
treeModel = DecisionTreeClassifier()
treeModel=treeModel.fit(X_train, y_train)
```

```python
DecisionTreeClassifier()
```

```python
plt.figure(figsize=(15, 10))
tree.plot_tree(treeModel, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
y_pred = treeModel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
